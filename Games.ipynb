## Подготовка данных ##

### Приведение к стилю и замена типа ###

```python
# привел колонки к стилю
games.columns = games.columns.str.lower()
# меняю тип данных на целочисленный
games['year_of_release'] = games['year_of_release'].astype(pd.Int64Dtype()) 

# столбец с общей ценой
games['total_sales'] = games.filter(like='sales').sum(axis=1)
```

<div class="alert alert-info">
    
<font size="3"><b> Привел к стилю название столбцов. Поменял тип данных в столбце year_of_release 
на целочисленный для удобства представления. Создал столбец с общей ценой.</b></font>
    
</div>

### Обработка 'tbd' ###

```python
# срез по значениям
games_tbd = games.query('user_score == "tbd"') 

display(f'Количество значений с tbd: {len(games_tbd)}') # сколько всего значений
display(f'Общая выручка по tbd: {games_tbd["total_sales"].sum()}') # общая выручка по tbd
display(f'Общая выручка по всем играм: {games["total_sales"].sum()}') # общая выручка по всем играм
games['user_score'] = pd.to_numeric(games['user_score'], errors='coerce') # так как при выводе
# ошибка, перевел все в вещественный тип
display(games_tbd.head())
display(games_tbd.info())
```

<div class="alert alert-info">
    
<font size="3"><b> 2424 позиции с рейтингом для уточнения. Посчитал общую выручку по ним и сравнил с общей выручкой по всем играм. Что составляет пять процентов. Можно предположить, что из-за невостребованности рейтинги этих игр не набрали достаточного количества голосов.</b></font>
    
</div>

```python
display(f'Максимальная оценка: {games["user_score"].max()}') # максимальная оценка
display(f'Минимальная оценка: {games["user_score"].min()}') # минимальная оценка
```

<div class="alert alert-info">
    
<font size="3"><b> Максимальный рейтинг 9.7, минимальный 0.

## Проверка гипотез ##

### Проверка гипотезы о средних пользовательских рейтингах платформ Xbox One и PC ###

<div class="alert alert-info">
<b> Рассматриваем гипотезу о равенстве средних двух генеральных совокупностей и альтернативную двухсторонюю о неравенстве.
    Для проверки используем уровень статистической значимости в 5 %. Как общепринятое пороговое значение.
    Н0: μ1 = μ2, Н1: μ1 != μ2,
    где Н0 - средние пользовательские оценки платформ XOne\PС равны,
    Н1 - средние пользовательские оценки платформ XOne\PС не равны.

```python
# делаем срез по платформам
games_forecast_xone = games_forecast.query('platform == "XOne"')
games_forecast_xone = games_forecast_xone.dropna(subset=['user_score'])
games_forecast_pc = games_forecast.query('platform == "PC"')
games_forecast_pc = games_forecast_pc.dropna(subset=['user_score'])

alpha = 0.05
# специальный метод
results = st.ttest_ind(
    games_forecast_xone['user_score'], 
    games_forecast_pc['user_score'], equal_var = False)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

<div class="alert alert-info">
    
<font size="3"><b>Гипотезу о равенстве не получилось отвергнуть. Вероятность получить такое или большее различие
случайно 15 %. Это явно большая вероятность, чтобы делать вывод о значимом различии между 
средними рейтингами.
</b></font>
