#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

## Обучение

### Функция для метрики

```python
def RMSE(target, predict):
    return (mean_squared_error(target, predict))**0.5

# метрика RMSE
RMSE_score = make_scorer(RMSE, greater_is_better=False)
```

### Функция создания признаков

```python
def make_features(data, max_lag, rolling_mean_size):
    data['hour'] = data.index.hour
    data['dayofweek'] = data.index.dayofweek
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    data = data.dropna()
```

```python
# применим функцию с лагом в 2 часа и скользящим средним 24
make_features(taxi_resample, 2, 24)
taxi_resample = taxi_resample.dropna()
taxi_resample.head()
```

### Признаки модели

```python
train, test = train_test_split(taxi_resample, shuffle=False, test_size=0.1)

x_train = train.drop('num_orders', axis=1)
y_train = train['num_orders']

x_test = test.drop('num_orders', axis=1)
y_test = test['num_orders']
```

```python
# размеры
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Обучение моделей

#### Линейная регрессия

```python
%%time
splits = TimeSeriesSplit(n_splits=8).split(x_train)
# модель
LR = LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=True)
# обучение
LR.fit(x_train, y_train)
# RMSE с помощью кросс-валидации
LR_cv_score = pd.Series(cross_val_score(LR, x_train, y_train, scoring=RMSE_score,cv=splits)).mean()*(-1)
print('Среднее качество линейной регрессии:', LR_cv_score)
```

#### Случайный лес

```python
%%time
splits = TimeSeriesSplit(n_splits=8).split(x_train)
#Объявляю модель
RFR = RandomForestRegressor(random_state=12345)
#Определяю словарь с набором параметров
RFR_parametrs = { 'n_estimators': range(5, 35, 5),
                     'max_depth': range(1, 20, 2),
                     'min_samples_leaf': range(1, 5),
                     'min_samples_split': range(2, 10, 2) }

#Применяю GridSearchCV с кросс-валидацией
RFR_grid = GridSearchCV(RFR, RFR_parametrs, scoring=RMSE_score, cv=splits)
RFR_grid.fit(x_train, y_train)
```

```python
%%time

RFR_params = RFR_grid.best_params_
RFR_score = RFR_grid.score(x_train, y_train) * (-1)
print('Лучшие параметры: ', RFR_params)
print('Лучшее качество: ',RFR_score)
```

### CatBoost

```python
%%time

splits = TimeSeriesSplit(n_splits=8).split(x_train)
#Объявляю модель
CB = cb.CatBoostRegressor()
#Определяю словарь с набором параметров
CB_parametrs = {'learning_rate': [0.02, 0.1],
                'depth': [4, 6, 8]
               }

#Применяю GridSearchCV с кросс-валидацией
CB_grid = GridSearchCV(CB, CB_parametrs, scoring=RMSE_score, cv=splits)
CB_grid.fit(x_train, y_train, verbose=100)
```

```python
%%time

CB_params = CB_grid.best_params_
CB_score = CB_grid.score(x_train, y_train) * (-1)
print('Лучшие параметры: ', CB_params)
print('Лучшее качество: ', CB_score)
```

## Тестирование

### Проверка на адекватность

```python
%%time

pred_adec = y_test.shift()
pred_adec.iloc[0] = y_train.iloc[-1]

print('RMSE модели на адекватность:', round(RMSE(y_test, pred_adec), 2))
```

### Тестирование на линейной регрессии.

```python
%%time

LR_test_model = LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=True)
LR_test_model.fit(x_train, y_train)
LR_test_model_predict = LR_test_model.predict(x_test)

LR_RMSE = round(RMSE(y_test, LR_test_model_predict), 2)
print('RMSE линейной регрессии:', LR_RMSE)
```

```python
# функция сравнения придсказаний и целевого признака
def comparison(predict):
    # смотрим, где тестовая
    test_date = pd.DataFrame({'target': y_test, 'predict': predict})
    test_date.plot()
    # вибираем из тестовой
    test_date['2018-08-15':'2018-08-18'].plot()
```

```python
# строим для линейной
comparison(LR_test_model_predict)
```

<font size="3"><b> Видим плохое попадание линейной регрессии вследствие запаздывания.
</b></font>
    
</div>

### Тестирование на случайном лесе

```python
%%time

RFR_test_model = RandomForestRegressor(random_state=12345,
                                     max_depth=17,
                                     min_samples_leaf = 4,
                                     min_samples_split = 2,
                                     n_estimators = 30)

RFR_test_model.fit(x_train, y_train)
RFR_test_model_predict = RFR_test_model.predict(x_test)

#Определяю метрику RMSE
RFR_RMSE = round(RMSE(y_test, RFR_test_model_predict), 2)
print('RMSE cлучайного леса:', RFR_RMSE)
```

```python
# строим для график сравнения для леса
comparison(RFR_test_model_predict)
```

<div class="alert alert-info">
