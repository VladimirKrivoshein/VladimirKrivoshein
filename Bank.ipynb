### Обучение моделей без учета дисбаланса

```python
# функция для обучения моделей
def learning_models(features_train, target_train, features_valid, target_valid):
    best_model_dt = None
    best_result_dt = 0
    best_depth_dt = 1
    for depth in range(1, 15):
        model = DecisionTreeClassifier(random_state=12345, max_depth=depth)# обучаю модель с заданной глубиной
        model.fit(features_train, target_train) # обучаю модель на тренировочной выборке
        prediction_valid = model.predict(features_valid)# получаю предсказания модели
        result = accuracy_score(target_valid, prediction_valid)# считаю качество модели на валидационной выборке
        if result > best_result_dt:
            best_model_dt = model # сохраняю наилучшую модель
            best_result_dt = result#  сохраняю наилучшее значение метрики accuracy на валидационных данных
            best_depth_dt = depth
    
    print('Лучшая глубина :', best_depth_dt)
    print('Лучшая точность для дерева решений :', best_result_dt)

    
    best_model_rf = None
    best_result_rf = 0
    best_n_estimators_rf = 1
    for est in range(1, 100):
        model = RandomForestClassifier(random_state=12345, n_estimators=est) # обучаю модель с заданным количеством деревьев
        model.fit(features_train, target_train) # обучаю модель на тренировочной выборке
        result = model.score(features_valid, target_valid) # считаю качество модели на валидационной выборке
        if result > best_result_rf:
            best_model_rf = model # сохраняю наилучшую модель
            best_result_rf = result#  сохраняю наилучшее значение метрики accuracy на валидационных данных
            best_n_estimators_rf = est
         
    print('Лучшeе количество деревьев :', best_n_estimators_rf)
    print('Лучшая точность для случайного леса:', best_result_rf)
    
    model_lr = LogisticRegression(random_state=12345, solver='liblinear')
    model_lr.fit(features_train, target_train)
    result_lr = model_lr.score(features_valid, target_valid) 

    print('Лучшая точность для логистической регрессии:', result_lr)
   
```

```python
learning_models(features_train, target_train, features_valid, target_valid)
```

и случайного леса.
</b></font>

## Борьба с дисбалансом

### Увеличение выборки

```python
# создаем функцию для увеличения положительных ответов в выборке и перемешаем функцией shuffle
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

features_upsampled, target_upsampled = upsample(features_train, target_train, 4)

# посмотрим распределение целевого признака
target_upsampled.value_counts(normalize = 1)

```

<div class="alert alert-info">
    
<font size="3"><b> Получен баланс классов целевых признаков с соотношением 1 : 1 для обучающей выборки.
    
</div>

```python
# расчитаем эмпирическим путем значения гиперпараметров
best_model_rf = None
best_result_rf = 0
best_n_estimators_rf = 1
best_depth_rf = 1
best_leaf_rf = 1
for est in range(1, 100):
    for depth in range(1, 15):
        model = RandomForestClassifier(random_state=12345, n_estimators=est,
                                           max_depth=depth, min_samples_leaf=50, n_jobs = -1,
                                           class_weight = 'balanced') # обучаю модель с заданным количеством деревьев
        model.fit(features_upsampled, target_upsampled)# обучаю модель на тренировочной выборке
        rf_prediction = model.predict(features_valid)
        result = recall_score(target_valid, rf_prediction) # считаю качество модели на валидационной выборке
        if (result + f1_score(target_valid, rf_prediction) + precision_score(target_valid, rf_prediction) + roc_auc_score(target_valid, rf_prediction)) > best_result_rf:
            best_model_rf = model # сохраняю наилучшую модель
            best_result_rf = result#  сохраняю наилучшее значение метрики accuracy на валидационных данных
            best_n_estimators_rf = est
            best_depth_rf = depth
            
print('Лучшeе количество деревьев :', best_n_estimators_rf)
print('Лучшая глубина для случайного леса:', best_depth_rf)
```
